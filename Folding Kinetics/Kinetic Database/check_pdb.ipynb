```python
import Bio.PDB as bpdb
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
kinetic_data = pd.read_pickle('kinetic_data.pkl')
kinetic_data.insert(5, "actual_length", 0, True)
kinetic_data.head()
```

```python
kinetic_data[kinetic_data['pdb'].str.startswith('1PGB')]
```

```python
io = bpdb.PDBIO()
for index, entry in kinetic_data.iterrows():
    struct = bpdb.PDBParser().get_structure('tmp', os.path.join('pdb_files', f'{entry.pdb}.pdb'))
    model = struct[0]
    kinetic_data.loc[index,'actual_length'] = len(list(model.get_residues()))
```
