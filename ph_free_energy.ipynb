## Net Charge

````python
df = net_charge.melt(ignore_index=False).query('variable != "1ev2"')
(
    ggplot(df, aes(x='df.index', y='value', colour='variable')) +
    geom_point() + geom_line() +
    ylab('Net Charge /eV') + xlab('pH')
)
````

## $\Delta Q_{\textit{binding}}$

````python
dQ_binding = pd.DataFrame()
dQ_binding.insert(0, 'FGF2-API5', net_charge['6l4o'] - net_charge['4oee'] - net_charge['3v6a'])
#dQ_binding.insert(0, 'FGF2-FGFR1', net_charge['1evt'] - net_charge['4oee'] - net_charge['5a4c'])
#dQ_binding.insert(0, 'Tetramer FGF2-FGFR2', net_charge['1ev2']/4 - net_charge['4oee'] - net_charge['1oec'])
dQ_binding.insert(0, 'FGF2-FGFR2', net_charge['1ev2-fgf2fgfr2'] - net_charge['1ev2-fgf2'] - net_charge['1ev2-fgfr2'])
dQ_binding
````

````python
df = dQ_binding.melt(ignore_index=False)
(
    ggplot(df, aes(x='df.index', y='value', colour='variable')) +
    geom_line() + geom_point() +
    ylab('$\Delta Q_{binding}$') + xlab('pH')
)
````

## $\Delta G_{\textit{binding}}$

````python
dG_binding = dQ_binding.copy()
for i in reversed(range(len(dG_binding)-1)):
    dG_binding.iloc[i] = dG_binding.iloc[i+1] + dG_binding.iloc[i]
dG_binding = -2.3 * dG_binding
dG_binding = dG_binding - dG_binding.min()
dG_binding
````

````python
df = dG_binding.melt(ignore_index=False)
(
    ggplot(df, aes(x='df.index', y='value', colour='variable')) +
    geom_line() + geom_point() +
    ylab('$\Delta G_{binding} /RT$') + xlab('pH')
)
````
