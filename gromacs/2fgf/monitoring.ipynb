```python
import numpy as np
import pandas as pd
from plotnine import *
```

## Energy Minimization

```python
step, potential = np.loadtxt('potential.xvg', comments=['#','@'], unpack=True)
em_potential = pd.DataFrame({'step': step, 'potential': potential})
```

```python
(
    ggplot(em_potential, aes(x='step', y='potential')) +
    geom_line() +
    ylab('Potential Energy (kJ mol^-1)')
)
```

## Equilibration NVT
### Temperature

```python
time, temperature = np.loadtxt('temperature.xvg', comments=['#','@'], unpack=True)
nvt_temperature = pd.DataFrame({'time': time, 'temperature': temperature})
```

```python
(
    ggplot(nvt_temperature, aes(x='time', y='temperature')) +
    geom_line() +
    ylab('Temperature (K)') + xlab('Time (ps)')
)
```

```python
time, density = np.loadtxt('density.xvg', comments=['#','@'], unpack=True)
time, pressure = np.loadtxt('pressure.xvg', comments=['#', '@'], unpack=True)
npt = pd.DataFrame({'time': time, 'density': density, 'pressure': pressure})
```

```python
(
    ggplot(npt, aes(x='time', y='pressure')) +
    geom_line()
)
```

```python
(
    ggplot(npt, aes(x='time', y='density')) +
    geom_line()
)
```

## MD Analysis

```python
time, rmsd = np.loadtxt('rmsd.xvg', comments=['#','@'], unpack=True)
time, rmsd_xtal = np.loadtxt('rmsd_xtal.xvg', comments=['#', '@'], unpack=True)
md_rmsd = pd.DataFrame({'time': time, 'rmsd': rmsd, 'rmsd_xtal': rmsd_xtal})
```

```python
(
    ggplot(md_rmsd.melt('time'), aes(x='time', y='value', colour='variable')) +
    geom_line()
)
```

```python
time, rg, rg_x, rg_y, rg_z = np.loadtxt('gyrate.xvg', comments=['#','@'], unpack=True)
md_rg = pd.DataFrame({'time': time, 'Rg': rg})
```

```python
(
    ggplot(md_rg.melt('time'), aes(x='time', y='value', colour='variable')) +
    geom_line()
)
```
